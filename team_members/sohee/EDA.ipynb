```python
import olefile
import pandas as pd

# CSV 다시 읽기
df = pd.read_csv('data/raw/data_list.csv')
df['텍스트_길이'] = df['텍스트'].str.len()

# 다양한 길이의 샘플 선택
samples = {
    '최소': df.loc[df['텍스트_길이'].idxmin()],
    '25%': df.loc[(df['텍스트_길이'] - df['텍스트_길이'].quantile(0.25)).abs().idxmin()],
    '중앙값': df.loc[(df['텍스트_길이'] - df['텍스트_길이'].median()).abs().idxmin()],
    '75%': df.loc[(df['텍스트_길이'] - df['텍스트_길이'].quantile(0.75)).abs().idxmin()],
    '최대': df.loc[df['텍스트_길이'].idxmax()]
}

print("샘플 문서 선정:")
print()
for name, doc in samples.items():
    print(f"{name:8s}: {doc['사업명'][:40]:40s} (CSV: {doc['텍스트_길이']:,}자)")

# HWP에서 텍스트 추출 함수
def extract_hwp_text(filepath):
    try:
        hwp = olefile.OleFileIO(filepath)
        if hwp.exists('PrvText'):
            prv_text = hwp.openstream('PrvText').read()
            text = prv_text.decode('utf-16le', errors='ignore')
            text = text.replace('\x00', '')
            hwp.close()
            return text
        else:
            hwp.close()
            return None
    except Exception as e:
        return f"Error: {e}"

# 각 샘플 문서 추출
print("\n" + "="*80)
print("HWP 텍스트 추출 결과:")
print("="*80)

results = []
for name, doc in samples.items():
    file_path = f"data/raw/files/{doc['파일명']}"
    hwp_text = extract_hwp_text(file_path)
    
    if hwp_text and isinstance(hwp_text, str):
        hwp_len = len(hwp_text)
        csv_len = doc['텍스트_길이']
        ratio = hwp_len / csv_len if csv_len > 0 else 0
        
        results.append({
            '분류': name,
            'CSV 길이': csv_len,
            'HWP 길이': hwp_len,
            '비율': f"{ratio:.1f}배",
            '차이': hwp_len - csv_len
        })
        
        print(f"\n[{name}] {doc['사업명'][:50]}")
        print(f"  CSV: {csv_len:,}자")
        print(f"  HWP: {hwp_len:,}자")
        print(f"  비율: {ratio:.1f}배 (차이: {hwp_len - csv_len:,}자)")
    else:
        print(f"\n[{name}] ❌ 추출 실패: {hwp_text}")

# 결과 요약 테이블
print()
print("결과 요약:")
print()
result_df = pd.DataFrame(results)
print(result_df.to_string(index=False))
```

```python
import olefile
import zlib

def extract_bodytext_with_olefile(hwp_path):
    """olefile로 HWP BodyText 추출"""
    try:
        hwp = olefile.OleFileIO(hwp_path)
        
        # BodyText 섹션들 찾기
        sections = []
        for entry in hwp.listdir():
            if entry[0] == 'BodyText' and entry[1].startswith('Section'):
                sections.append(entry)
        
        full_text = []
        
        for section in sorted(sections):
            try:
                stream = hwp.openstream(section)
                raw_data = stream.read()
                
                # 압축 해제 시도
                try:
                    decompressed = zlib.decompress(raw_data, -15)
                    text = decompressed.decode('utf-16', errors='ignore')
                except:
                    text = raw_data.decode('utf-16', errors='ignore')
                
                # NULL 문자와 제어 문자 제거
                text = ''.join(c for c in text if c.isprintable() or c in ['\n', '\t', ' '])
                full_text.append(text)
                
            except Exception as e:
                print(f"섹션 {section} 처리 중 에러: {e}")
                continue
        
        hwp.close()
        return '\n'.join(full_text)
        
    except Exception as e:
        print(f"파일 열기 실패: {e}")
        return None

# 실행!
bodytext = extract_bodytext_with_olefile(test_path)

if bodytext:
    print(f"추출된 텍스트 길이: {len(bodytext)}자")
    print(f"\n처음 1000자:\n{bodytext[:1000]}")
else:
    print("추출 실패")
```

```python
import olefile
import re

def extract_text_improved(hwp_path):
    """개선된 HWP 텍스트 추출"""
    try:
        hwp = olefile.OleFileIO(hwp_path)
        
        # 1. PrvText 먼저 시도 (미리보기)
        if hwp.exists('PrvText'):
            prv_stream = hwp.openstream('PrvText')
            prv_data = prv_stream.read()
            
            # UTF-16 LE 디코딩
            prv_text = prv_data.decode('utf-16le', errors='ignore')
            
            # NULL 문자 및 불필요한 문자 제거
            prv_text = prv_text.replace('\x00', '')
            prv_text = re.sub(r'[\x00-\x08\x0b\x0c\x0e-\x1f]', '', prv_text)
            
            print(f"✅ PrvText 추출 성공: {len(prv_text)}자")
        
        # 2. BodyText 시도 (더 완전한 내용)
        body_texts = []
        
        # BodyText 섹션들 찾기
        for entry in hwp.listdir():
            if len(entry) >= 2 and entry[0] == 'BodyText':
                if entry[1].startswith('Section'):
                    try:
                        stream = hwp.openstream(entry)
                        data = stream.read()
                        
                        # 여러 인코딩 시도
                        for encoding in ['utf-16le', 'utf-16', 'cp949', 'euc-kr']:
                            try:
                                text = data.decode(encoding, errors='ignore')
                                # 유효한 텍스트인지 확인 (한글이 포함되어 있는지)
                                if re.search(r'[가-힣]', text):
                                    # NULL 및 제어 문자 제거
                                    text = text.replace('\x00', '')
                                    text = re.sub(r'[\x00-\x08\x0b\x0c\x0e-\x1f]', '', text)
                                    body_texts.append(text)
                                    break
                            except:
                                continue
                    except Exception as e:
                        print(f"섹션 {entry} 처리 중 에러: {e}")
                        continue
        
        hwp.close()
        
        # 결과 합치기
        result = {
            'preview': prv_text if 'prv_text' in locals() else '',
            'body': '\n'.join(body_texts) if body_texts else '',
            'total_length': len(prv_text if 'prv_text' in locals() else '') + len('\n'.join(body_texts) if body_texts else '')
        }
        
        return result
        
    except Exception as e:
        print(f"❌ 전체 처리 실패: {e}")
        return None

# 테스트 실행
result = extract_text_improved(test_path)

if result:
    print("\n" + "="*50)
    print("추출 결과:")
    print("="*50)
    print(f"Preview 텍스트: {len(result['preview'])}자")
    print(f"Body 텍스트: {len(result['body'])}자")
    print(f"총 길이: {result['total_length']}자")
    
    print("\n처음 500자 미리보기:")
    print(result['preview'][:500] if result['preview'] else result['body'][:500])
```
