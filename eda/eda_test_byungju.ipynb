```python
import pandas as pd

# CSV 파일 읽기
try:
    df = pd.read_excel('./data/raw/data_list.xlsx')
except:
    df = pd.read_csv('./data/raw/data_list.csv')

print(f"전체 데이터 개수: {df.shape[0]}개")
print("컬럼 목록:", df.columns.tolist())

# 상위 5개만 눈으로 확인
df.head()
```

```python
import os

# 실제 폴더에 있는 파일 이름 다 가져오기
real_files = os.listdir('./data/raw/files')
print(f"실제 파일 개수: {len(real_files)}개")

# 데이터프레임에 있는 파일명과 비교해보기
if '파일명' in df.columns:
    missing_files = []
    
    for name in df['파일명']:
        if name not in real_files:
            missing_files.append(name)
            
    if len(missing_files) == 0:
        print("완벽합니다! 목록에 있는 파일이 폴더에도 전부 있습니다.")
    else:
        print(f"주의! 목록에는 있는데 실제로는 없는 파일이 {len(missing_files)}개 있습니다.")
        print("없는 파일 예시:", missing_files[:3])
else:
    print("'파일명'이라는 컬럼을 못 찾겠어요. df.columns를 보고 정확한 컬럼명을 넣어주세요.")
```
